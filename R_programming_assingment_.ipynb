```python
import random
import pandas as pd
import numpy as np
# Create a file that contains 1000 lines of random strings.


with open('random_strings.txt', 'w') as f:
  for _ in range(1000):
    string = ''.join(random.choices('abcdefghijklmnopqrstuvwxyz', k=10))
    f.write(string + '\n')

# Q2) Create a random dataset of 100 rows and 30 columns . All the values are defined between [1,200]


# Create a dataframe with 100 rows and 30 columns
df = pd.DataFrame(np.random.randint(1, 201, size=(100, 30)))

# Display the dataframe
print(df.head())

# (1) Replace all the values with NA in the dataset defined between [10,60].Print the count of number rows having missing values.
df[(df >= 10) & (df <= 60)] = np.nan

# Print the count of number rows having missing values
print(df.isnull().sum().sum())
```

(2) Replace all the NA values with average of the columns value.

```python
df.fillna(df.mean(), inplace=True)
```

(3) Find the Pearson correlation among all the columns and plot heat map. Also select those columns having correlation <=0.7

```python
# prompt: Find the Pearson correlation among all the columns and plot heat map. Also select those columns having correlation <=0.7

import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the Pearson correlation matrix
corr = df.corr()

# Plot the heatmap
sns.heatmap(corr, vmin=-1, vmax=1, annot=True, cmap='rocket_r')
plt.show()

# Select columns with correlation <= 0.7
selected_cols = [col for col in corr.columns if abs(corr[col]).max() <= 0.7]

# Print the selected columns
print(selected_cols)
```

Q3) Create a random dataset of 500 rows and 10 columns.
Columns 1 to 4 are defined between [-10,10];
Columns 5 to 8 are defined between [10,20];
Columns 9 to 10 are defined between [-100,100].
Apply following clustering algorithms ; determine the optimal number of clusters and plot distance metric graph using each algorithm.
(1) k-mean clustering   (2) Hierarchical clustering

```python
import numpy as np
import pandas as pd

np.random.seed(42)
data = np.hstack([
    np.random.uniform(-10, 10, (500, 4)),  # Columns 1 to 4
    np.random.uniform(10, 20, (500, 4)),   # Columns 5 to 8
    np.random.uniform(-100, 100, (500, 2)) # Columns 9 to 10
])
df = pd.DataFrame(data, columns=[f'col{i+1}' for i in range(10)])

```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

wcss = []
range_n_clusters = range(1, 11)
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, wcss, marker='o', linestyle='-', color='b')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
plt.title('Elbow Method For Optimal Number of Clusters')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

kmeans_silhouette_scores = []
range_n_clusters = range(2, 11)
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(df)
    silhouette_avg = silhouette_score(df, cluster_labels)
    kmeans_silhouette_scores.append(silhouette_avg)

plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, kmeans_silhouette_scores, marker='o', linestyle='-', color='b')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores For K-means Clustering')
plt.grid(True)
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(df, method='ward')

plt.figure(figsize=(12, 8))
dendrogram(Z)
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.title('Dendrogram For Hierarchical Clustering')
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

hierarchical_silhouette_scores = []
for n_clusters in range_n_clusters:
    cluster_labels = fcluster(Z, n_clusters, criterion='maxclust')
    silhouette_avg = silhouette_score(df, cluster_labels)
    hierarchical_silhouette_scores.append(silhouette_avg)

plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, hierarchical_silhouette_scores, marker='o', linestyle='-', color='b')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores For Hierarchical Clustering')
plt.grid(True)
plt.show()
```

Q4) Create a random dataset of 600 rows and 15 columns. All the values are defined between [-100,100]. PErform tha following operations:

(3) Plot the Box plot of each column in single graph.

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a random dataset of 600 rows and 15 columns
df = pd.DataFrame(np.random.randint(-100, 101, size=(600, 15)))

# Plot scatter graph between Column 5 and columnn 6
plt.scatter(df.iloc[:, 5], df.iloc[:, 6])
plt.xlabel('Column 5')
plt.ylabel('Column 6')
plt.title('Scatter Plot of Column 5 vs Column 6')
plt.show()

# Plot histogram of each column in single graph
df.hist(figsize=(10, 8), bins=10)
plt.suptitle('Histograms of All Columns')
plt.show()

# Plot the Box plot of each column in single graph
df.boxplot(figsize=(10, 8))
plt.suptitle('Box Plots of All Columns')
plt.show()
```
